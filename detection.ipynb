```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
raw_mail_data=pd.read_csv("mail_data.csv")
```

```python
print(raw_mail_data)
```

```python
# replace the null values with a null string
mail_data=raw_mail_data.where((pd.notnull(raw_mail_data)),'')
```

```python
mail_data.head()
```

```python
mail_data.shape
```

```python
# label encoding: label spam mail as 0; and ham mail as 1;
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()

```

```python
data=pd.DataFrame(mail_data)
```

```python
data['Category']=encoder.fit_transform(data['Category'])
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
# separating the data as texts and label
x=data['Message']

y=data['Category']
```

```python
x
```

```python
print(y)
```

### splitting the data into training and test data



```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=3)
```

```python
print(x.shape)
print(x_train.shape)
print(x_test.shape)
```

### Feature Extraction

```python
# transorm the text data to feature vectors that can be used as input to the logistic regression
feature_extraction=TfidfVectorizer(min_df=1,stop_words='english',lowercase=True)
x_train_features=feature_extraction.fit_transform(x_train)
x_test_features=feature_extraction.transform(x_test)

#convert y_train and y_test values as integers
y_train=y_train.astype('int')
y_test=y_test.astype('int')
```

```python
print(x_train_features)
```

### trainniing the model

```python
model=LogisticRegression()
```

```python
# training the logistic regression model with the training data
model.fit(x_train_features,y_train)
```

### evaluting the trained model

```python
# preidction on training data
prediction_on_training_data=model.predict(x_train_features)
```

```python
accuracy_on_training_data=accuracy_score(y_train,prediction_on_training_data)
```

```python
print('Accuracy on training data :',accuracy_on_training_data)
```

```python
# preidction on testing data
prediction_on_testing_data=model.predict(x_test_features)
```

```python
accuracy_on_testing_data=accuracy_score(y_test,prediction_on_testing_data)
```

```python
print('Accuracy on testing data :',accuracy_on_testing_data)
```

### Building a Predictive System

```python
input_mail=["Eh u remember how 2 spell his name... Yes i did. He v naughty make until i v wet."]
# convert text to feature vectors
input_data_features=feature_extraction.transform(input_mail)

# making prediction
prediction=model.predict(input_data_features)
print(prediction)
if (prediction[0]==0):
    print("Ham Mail")
else:
    print("Spam Mail")
```
